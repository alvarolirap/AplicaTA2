## Eliminacion de los signos de puntuacion

Se eliminan los signos de puntuacion por que no contribuyen con el objetivo y crea palabras diferentes que no deberian serlo.
Ejemplo: Audi / Audi's / Audi: / Audi,

```python
import re
from nltk.corpus import stopwords

pattern = r"[&$/%'()+,-.:;<=>?[\\]^_`{|}~!]"
    
def borrar_puntuacion(row):
    row.text = re.sub(pattern,"",row.text)
    return row.text
    
train_data.text = train_data.apply(borrar_puntuacion, axis=1)

train_data
```

## Convertir el texto a minusculas

 Se pasa todo a minúscula porque para esta tarea no importa si las palabras estan en mayúsculas o minúsculas

```python
def to_lower(row):
    row[1] = row[1].lower()
    return row[1]
    
train_data.text = train_data.apply(to_lower, axis=1)

train_data
```

## Limpiar la data

```python
import re
from nltk.corpus import stopwords

pattern = r"\bthx\b "
pattern2 = r"\bhttp.*\b "
pattern3 = r"\blol\b |\blel\b |\brofl\b |\broflmao\b |\blmao\b |\blmfao\b |\b[he]+]\b |\b[ja]+\b |\b[ha]+\b "
pattern4 = r"\bwtf\b |\bwafuq\b |\btfck\b |\bthefuck\b "
pattern5 = r"\byass\b |\byep\b |\byei\b |\byay\b |\byeah\b "
pattern6 = r"\baf\b "
pattern7 = r"\bppl\b "
pattern8 = r"\blit\b "
pattern9 = r"\bbae\b "
pattern10 = r"\bgotcha\b "
pattern11 = r"\bwaddup\b |\bwassup\b |\bwadup\b "
pattern12 = r"\bgwanna\b "
pattern13 = r"'ve\b "
pattern14 = r"n't\b "
pattern15 = r"'d\b "
pattern16 = r"\bi'm\b "
pattern17 = r"\bwhat's\b"
pattern18 = r"\bit's\b"
  
def replace_slang(row):
    row[1] = re.sub(pattern," thanks ",row[1])
    row[1] = re.sub(pattern2," ",row[1])
    row[1] = re.sub(pattern3," haha ",row[1])
    row[1] = re.sub(pattern4," what the fuck ",row[1])
    row[1] = re.sub(pattern5," yes ",row[1])
    row[1] = re.sub(pattern6," as fuck ",row[1])
    row[1] = re.sub(pattern7," people ",row[1])
    row[1] = re.sub(pattern8," lit ",row[1])
    row[1] = re.sub(pattern9," babe ",row[1])
    row[1] = re.sub(pattern10," got you ",row[1])
    row[1] = re.sub(pattern11," what is up ",row[1])
    row[1] = re.sub(pattern12," want to ",row[1])
    row[1] = re.sub(pattern13," have ",row[1])
    row[1] = re.sub(pattern14," not ",row[1])
    row[1] = re.sub(pattern15," had ",row[1])
    row[1] = re.sub(pattern16," i am ",row[1])
    row[1] = re.sub(pattern17," what is ",row[1]) 
    row[1] = re.sub(pattern18," it is ",row[1])
    
    return row[1]

train_data.text = train_data.apply(replace_slang, axis=1)
train_data
```

## Remover Stopwords

Se remueven las 'stopwords' ya que son terminos muy comunes que no tienen suficiente relevancia para el objetivo de distinguir entre una opinion a favor o en contra

```python
stop = set(stopwords.words('english'))

stopwords_removed = 0

for index, row in train_data.iterrows():
    for i in row[1].split(" "):
        if i in stop:
            stopwords_removed = stopwords_removed + 1

train_data.text = train_data.text.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

print ("Numero de stopwords eliminados: ",stopwords_removed)

train_data
```

## Stemming

Se aplica Stemming porque al ser una red social, las personas hablan de manera informal, al reducir las palabras a su raiz podemos procesarlas de mejor manera

```python

data_stem=train_data.copy()
data_stem["text"] = data_stem["text"].map(lambda x:x.split())
```

```python
porter = nltk.PorterStemmer()
data_stem['text'] = data_stem["text"].map(lambda x: [porter.stem(y) for y in x])
data_stem
```

## Lemmatize

```python
wnl = nltk.WordNetLemmatizer()
data_lem=train_data.copy()
data_lem["text"] = data_lem["text"].map(lambda x:x.split())
```

```python
data_lem["text"] = data_lem["text"].map(lambda x:[wnl.lemmatize(t) for t in x])
```

```python
sentences = [x.split() for x in train_data['text']]
```

```python
num_features = 100
min_word_count = 25
num_workers = 4
context= 6
downsampling = 1e-3
```

```python
import gensim
from gensim.models import Word2Vec
model = Word2Vec(sentences, workers=num_workers, size=num_features,
             min_count=min_word_count, window=context, sample=downsampling, iter=10)
```

```python
def vectorize(model, X):
    return np.array([
        np.mean([model[w] for w in words if w in model] or [np.zeros(self.dim)], axis=0) for words in X
    ])
```

```python
X_word2vec = vectorize(model, train_data['text'])
```
